```python
def listToString(s):  
#Literally just copy-pasted this from geeks2geeks lol
    str1 = ""  
  
    for ele in s:  
        str1 += ele + " "
     
    return str1  
```

```python
def convertToPOS(text):
#Converts text to POS using spacy. This particular method ignores punctuation and spaces

	posText = ""
	nlp = spacy.load("en_core_web_sm")
	doc = nlp(str(text))

	for token in doc:

		if not(token.pos_ == 'SPACE' or token.pos_ == 'PUNCT'):

			if token.text == '.':
				posText += "."
			elif posText == ',':
				posText += ","
			else:
				posText = posText + " " + token.pos_ 
		
	return posText
```

```python
def identifyMostCommonGrams(text, N, X):

	mostFrequentGrams = np.empty(X)

	amountOfNGrams = len(text.split()) - (N-1)
	gramArray = ['','']
	individualWordsList = text.split()
	gramWordsList = []
	for i in range(amountOfNGrams):
		gramWordsList.append(0)
	gramFrequencyList = []
	for i in range(amountOfNGrams):
		gramFrequencyList.append(0)

	#print(individualWordsList)

	for q in range(len(individualWordsList)-(N-1)):

		temp = ""

		for i in range(N):

			temp = temp + individualWordsList[i + q] + ' '

		gramWordsList[q] = temp

	#print(gramWordsList)
	
	
		p = Counter(gramWordsList)
		#print(p.most_common(3))
		j = p.most_common(X)
	finalList = []
	for m in range(X):
		finalList.append(j[m][0])
	return finalList
```

```python
def uniqueWordRatio(text):
	#This finds the percentage of unique words to total words. 
	#For example, if there are 5 words and 4 of them are unique (1 is used twice), then the percentage is 4/5 = 80%, or 0.80. 
	#1.00 means all words are unique, 0 means none are. 
	#Logically then, the greater the percentage, the more unique words the person used, thus representing a wider vocabulary
	totalWordsList = text.split()
	uniqueWordsList = []
	totalNumberOfWords = 0
	numberOfUniqueWords = 0
	#progressiveCountUniqueWords = []
	#progressiveCountTotalWords = []
	for element in totalWordsList:
		if not element in uniqueWordsList:
			uniqueWordsList.append(element)
			numberOfUniqueWords += 1
	#	progressiveCountUniqueWords.append(numberOfUniqueWords)
		totalNumberOfWords += 1
	#	progressiveCountTotalWords.append(totalNumberOfWords)
	
	#plt.axis([0, 1500, 0, 1500])
	#plt.autoscale(enable=True, axis=u'both', tight=False)
	#plt.plot(np.array(progressiveCountTotalWords), np.array(progressiveCountUniqueWords))
	#print(progressiveCountUniqueWords)
	#plt.ylabel('Unique Words')
	#plt.xlabel('Total amount of Words')
	#plt.show()
	return Decimal(str(numberOfUniqueWords)) / Decimal(str(len(totalWordsList)))
```

```python
def findAverageWordLength(writing):
	letterCount = 0
	wordCount = len(writing.split())
	for char in writing:
		if char in 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ':
			letterCount += 1
			#print(char)
	return Decimal(str(letterCount)) / Decimal(str(wordCount))
```

```python
def findMostPopularStopWords(text, X):
	stop_words = set(stopwords.words("english"))
	wordList = word_tokenize(text)
	stopWordList = [w for w in wordList if w in stop_words]
	
	#print(stopWordList)
	stopWordsText = listToString(stopWordList)
	#print(stopWordsText)
	
	mostPopularStopWords = identifyMostCommonGrams(stopWordsText, 1, X)
	
	return mostPopularStopWords
```

```python
def stopWordRatioToTotalWords(text):
	stop_words = set(stopwords.words('english'))
	totalNumberOfWords = len(text.split())
	wordList = word_tokenize(text)
	numberOfStopWords = 0
	for token in wordList:
		if token in stop_words:
			numberOfStopWords += 1
	return Decimal(str(numberOfStopWords)) / Decimal(str(totalNumberOfWords))
```
